```python
df['Cumilative Sales'] = df['Sales'].cumsum()
```

```python
Yearwise_Sales =  pd.DataFrame(df.groupby('Year')['Sales'].sum())

Yearwise_Sales
```

```python
Yearwise_Sales['YoY Diffrence'] = Yearwise_Sales['Sales'].iloc[0:].diff()

Yearwise_Sales
```

```python
Yearwise_Sales['YoY Growth'] = (Yearwise_Sales['YoY Diffrence']/Yearwise_Sales['Sales'])*100
Yearwise_Sales['YoY Diffrence'] = Yearwise_Sales['YoY Diffrence'].fillna(0)
Yearwise_Sales['YoY Growth'] = Yearwise_Sales['YoY Growth'].fillna(0)
```
